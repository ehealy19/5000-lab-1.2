```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'age': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    'height': [150, 155, 160, 165, 170, 175, 180, 185, 190, 195, 200, 205]},
    index = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6', 'p7', 'p8', 'p9', 'p10', 'p11', 'p12'])

plot_1=sns.relplot(
    data=df, 
    x="age", 
    y="height", 
    s=200
)

FS = 14
plot_1.set_axis_labels('Age of person (yrs)', 'Height (cm)', fontsize=FS)
plot_1.set_titles('Height versus Age', fontsize=FS)

plt.show()
```
